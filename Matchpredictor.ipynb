```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(
    n_estimators=200,
    min_samples_split=50,
    random_state=1,
    n_jobs=-1
)
```

```python
test = matches[matches["MatchDate"] >= '2024-01-01']
```

```python
predictors = [
    "home_code",
    "away_code",
    "HomeCorners",
    "AwayCorners",
    "HomeFouls",
    "AwayFouls",
    "HomeYellowCards",
    "AwayYellowCards"
]
```

```python
rf.fit(train[predictors], train["Target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
```

```python
acc = accuracy_score(test["Target"], preds)
```

```python
def rolling_averages(group, cols, new_cols):
    group = group.sort_values("MatchDate")
    rolling_stats = group[cols].rolling(3, closed='left').mean()
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)
    return group
```

```python
cols = ["FullTimeHomeGoals" ,"FullTimeAwayGoals" , "HomeShots", "AwayShots", "HomeShotsOnTarget" ,"AwayShotsOnTarget"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
matches_rolling = matches.groupby("HomeTeam").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel("HomeTeam")
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["MatchDate"] < "2024-01-01"]
    test  = data[data["MatchDate"] >= "2024-01-01"]

    rf.fit(train[predictors], train["Target"])
    preds = rf.predict(test[predictors])

    combined = pd.DataFrame({
        "actual": test["Target"],
        "prediction": preds
    })

    precision = precision_score(test["Target"], preds)
    return combined, precision
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_cols)
```
